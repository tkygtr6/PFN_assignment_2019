```python
wait1 = list(np.loadtxt("wait1_2", delimiter=" "))
low_value1 = []
high_value1 = []

for wait in wait1:
    if wait[0]:
        high_value1.append(wait[1])
    else:
        low_value1.append(wait[1])
print(sum(low_value1), sum(high_value1))
plt.hist(high_value1)
```

```python
wait_ = list(np.loadtxt("wait2_1", delimiter=" "))
low_value = []
high_value = []

for wait in wait_:
    if wait[0]:
        high_value.append(wait[1])
    else:
        low_value.append(wait[1])
        
range_min = min(low_value + high_value)
range_min = max(low_value + high_value)
print("sum", sum(low_value), sum(high_value))
print("avg", '{:.3g}'.format(sum(low_value)/len(low_value)))

plt.xlim([-1,26])
plt.title("No Priority (2-1)")
plt.xlabel('Wait Time [s]')
plt.ylabel('Count')
plt.hist(low_value, bins=int(max(low_value)), rwidth=0.8, label="Low")
# plt.hist(high_value, bins=int(max(high_value)), rwidth= 0.6, label="High")
plt.legend()
plt.savefig( 'wait2_1.pdf' )
```

```python
wait_ = list(np.loadtxt("wait2_2", delimiter=" "))
low_value = []
high_value = []

for wait in wait_:
    if wait[0]:
        high_value.append(wait[1])
    else:
        low_value.append(wait[1])
        
range_min = min(low_value + high_value)
range_min = max(low_value + high_value)
print("sum", sum(low_value), sum(high_value))
print("avg", '{:.3g}'.format(sum(low_value)/len(low_value)))
print("avg", '{:.3g}'.format(sum(high_value)/len(high_value)))

plt.xlim([-1,26])
plt.title("Naive Scheduling (2-2)")
plt.xlabel('Wait Time [s]')
plt.ylabel('Count')
plt.hist(low_value, bins=int(max(low_value)), rwidth=0.8, label="Low")
plt.hist(high_value, bins=int(max(high_value)), rwidth= 0.6, label="High")
plt.legend()
plt.savefig( 'wait2_2.pdf' )
```
